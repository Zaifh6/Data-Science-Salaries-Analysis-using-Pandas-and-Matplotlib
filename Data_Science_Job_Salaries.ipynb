## Exploratory Data Analysis : Data Science Job Salaries

###This dataset provides information about the salaries within the field of data science jobs. Our goal is to conduct an exploratory data analysis on this dataset, aiming to reveal intriguing insights and patterns hidden within the data. By delving into this analysis, we hope to gain a better understanding of the salary trends in the data science industry and potentially discover factors that contribute to variations in salaries.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/ds_salaries.csv')
df
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
df.columns
```

```python
df = df.drop('Unnamed: 0', axis=1)
df.head(10)
```

```python
df.rename(columns={
    'work_year': 'Work Year',
    'experience_level': 'Experience Level',
    'employment_type': 'Employment Type',
    'job_title': 'Job Title',
    'salary': 'Salary',
    'salary_currency': 'Salary Currency',
    'salary_in_usd': 'Salary in USD',
    'employee_residence': 'Employee Residence',
    'remote_ratio': 'Remote Ratio',
    'company_location': 'Company Location',
    'company_size': 'Company Size'
}, inplace=True)
df.head(10)
```

###Let's Ask Some Questions From The Data

###How many people are working in each category?

```python
df['Job Title'].value_counts()
```

###Highest salary of each job title?

```python
new_df = df.groupby(['Job Title']).max()
new_df['Salary in USD']
```

```python
plt.figure(figsize=(14, 6))
plt.bar(new_df.index, new_df['Salary in USD'])
plt.xlabel('Job Title')
plt.ylabel('Salary in USD')
plt.title('Highest Salaries by Job Title')
plt.xticks(rotation=90)


plt.tight_layout()
plt.show()
```

###What is the average salary in USD?

```python
new_df1 = df.groupby(['Job Title']).mean()
new_df1['Salary in USD']
```

```python
plt.figure(figsize=(14, 6))
plt.bar(new_df1.index, new_df1['Salary in USD'])
plt.xlabel('Job Title')
plt.ylabel('Average Salary in USD')
plt.title('Highest Average Salaries by Job Title')
plt.xticks(rotation=90)


plt.tight_layout()
plt.show()
```

###What is the highest salary in the dataset?

```python
df['Salary in USD'].max()
```

###What is the distribution of remote work ratios?

```python
df.head(10)
```

```python
# Count the number of remote and non-remote jobs
remote_counts = df['Remote Ratio'].value_counts()

# Create a pie chart
labels = ['Non-Remote', 'Remote']
sizes = [remote_counts.get(0, 0), remote_counts.get(100, 0)]
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  # Explode the 1st slice (i.e., 'Non-Remote')

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Distribution of Remote vs. Non-Remote Jobs')
plt.show()
```
